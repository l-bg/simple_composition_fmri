# An fMRI study of composition in noun and verb phrases
### Bonnasse-Gahot, Bemis, Perez-Guevara, Dehaene, Pallier
Compute contrasts for all subjects, and group, for the incremental model

```python
import numpy as np
import os.path as op
from joblib import load
import nibabel as nib

import simpcomp as sc

output_path = op.join(sc.model_incremental_folder, 'contrasts')
sc.make_dir(output_path)
```

```python
for sub_id in sc.subject_list:
    sub_glm_filename = op.join(sc.model_incremental_folder, 'glm', 'glm_{}'.format(sub_id))
    fmri_glm = load(sub_glm_filename)

    design_matrix = fmri_glm.design_matrices_[0]
    contrast_matrix = np.eye(design_matrix.shape[1])
    contrasts_basic = dict([(column, contrast_matrix[i])
                      for i, column in enumerate(design_matrix.columns)])

    contrasts = {}
    contrasts['resp'] = contrasts_basic['resp_r'] - contrasts_basic['resp_l']
   
    #####main effects    
    
    # structure
    contrasts['structure-verb'] = (contrasts_basic['wvp']
                                   -contrasts_basic['wvl'])
    contrasts['structure-noun'] = (contrasts_basic['wnp']
                                   -contrasts_basic['wnl']) 
    contrasts['structure'] = (contrasts['structure-verb'] 
                              + contrasts['structure-noun'])

    # type
    contrasts['type-phrase'] = (contrasts_basic['wvp']
                                -contrasts_basic['wnp'])
    contrasts['type-list'] = (contrasts_basic['wvl']
                              -contrasts_basic['wnl']) 
    contrasts['type'] = (contrasts['type-phrase'] 
                         + contrasts['type-list'])
    
    ##### 2-way interactions
    contrasts['structureXtype'] = (contrasts['structure-verb']
                                   - contrasts['structure-noun'])
    
    for name, val in contrasts.items():
        z_map = fmri_glm.compute_contrast(val, output_type='z_score')
        eff_map = fmri_glm.compute_contrast(val, output_type='effect_size')
        nib.save(z_map, op.join(output_path, '{}_{}_zmap.nii.gz'.format(name, sub_id)))
        nib.save(eff_map, op.join(output_path, '{}_{}_effsize.nii.gz'.format(name, sub_id)))
```

```python
# group analysis
for contrast in contrasts:
    maps = [op.join(output_path, '{}_{}_effsize.nii.gz'.format(contrast,sub_tmp)) for sub_tmp in sc.subject_list]   
    if maps == []:
        print("%s : no file with this mask" % mask)
    else:
        sc.create_one_sample_t_test(contrast, maps, output_path)  
```
